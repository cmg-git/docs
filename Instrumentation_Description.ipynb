# Instrumentation Description

The instrumentation used for the provided measuremnts consists of a set of sensors and measuring devices capable of receiving and transmiting information remotely. This report gives a description of the following measuring devices used:

1. Eco-Touch® Device
2. Indoor Temperature and Humidity Sensor
3. Outdoor Temperature and Humidity Sensor
3. Iluminance Sensor
4. Window Opening Detector
4. DHW Volumetric Meter

## Eco-Touch® Device

The main measuring device is the Eco-Touch® device by the company OGGA, an energy manager supplied with 230V
upstream of a circuit breaker. It equips the homes with a technical base to optimize and monitor the consumption of the energy. It makes the housing remotely controllable and alerts in the event of a malfunction. 


This instrument measures the electricity consumption, and receives and transmits information from other sensors. Therefore, the measurements can be:
1. Direct measurments - measured directly by the Eco-Touch®
2. Indirect measurments - measured by sensors and transmited to the Eco-Touch®

### Direct measuremnts
This Eco-Touch® instrument is connected to the main electrical board of the house measuring the electricity consumption on 4 toroids (current clamps). In this way, the electrical consumption for different use is directly measured by the Eco-Touch®.



When used for monitoring dwelings, the 4 current measurement toroids from the Eco-Touch® are installed as follows:

1. One will be set up on the general departure of the electrical meter from the accommodation.
2. One will be placed on the outlets circuit breaker.
3. One will be placed on the additional electrical DHW circuit.
4. One will be placed on the electric radiator circuit and air conditioning circuit (if present).

*Note: If no air conditioning or electric DHW is provided, the toroid (s) will be left on standby in the electrical panel in the closed position.All the wires supplying the circuits corresponding to the toroid label must be put in it.
It is necessary to put only the phases and not the neutrals.*

*Note: This is the most common use of the Eco-Touch® device but other applications can also be found like for example the use of the Eco-Touch® for measuring the electrical consumption for underfloor heating from more phases.* 

### Indirect measuremnts
The information from other sensors can be received by a wireless transmission with the use of a EnOcean® radio protocol. The EnOcean® radio has an open and secure protocol, able to transmit and receive information when used in combination with many other international systems. It is an energy efficient system with a satisifed technological maturity.





#### Indoor Temperature and Humidity Sensor

These sensors can be positioned in every indoor space where the monitoring is needed. The sensor can be mounted on every indoor surface with screws or sticky bands. They communicate remotely with an Eco-Touch® so that real-time temperature and humidity information is received. The comunication is enabled through the EnOcean® radio protcol.

The temperature and the relative humidity of the room is checked every 100sec, and sent to the Eco-Touch® when a change of 0.5°C temperature variation occures, while for the humidity when a +/-2% humidity variation is checked.
To make sure the Temperature and Humidity Sensor is always alive, it will broadcast an “heartbeat” temperature and humidity message every 15 to 30min. 



#### Outdoor Temperature and Humidity Sensor 

The sensor reports a temperature variation of 0.6°C, while for the humidity a variation of +/-2% is registered. If the
values remain unchanged, a message control is sent approximately every 100 to 3000 seconds depending on the charging state of the sensor. The reported values are comunicated to the Eco-Touch® device via the EnOcean® radio protcol.

#### Iluminance Sensor

The illuminance sensor from the company Eltako is equipped with a small solar cell with a battery allowing this sensor to work without an external solar supply. The measured information from the sensor is reported and transmitted to an Eco-Touch® device by protocol radio.

This outdoor light sensor sends information to the Eco-Touch® when the illuminance goes above 300Lux and at every change of brightness over about 500 Lux within about 10 seconds. Additionally, there is a control message approximately every 100 seconds. The measurements are reported to the Eco-Touch® devices via the EnOcean® radio protcol.

#### Window Opening Detector

The window opening/colsing detector is a very discreet sensor mounted on the door/window frame reporting every opening. In this way information about the ventilaion periods are obtainend such as when the tenants ventilate their accommodation and on what
duration. 

The presented system is composed of a fixed element which is mounted on the door frame (or window) and a movable element fixed to the leaf. The signal is interrupted when the door is open and the two elements are no longer facing each other. The information is then transmitted to the Eco-Touch®.

#### DHW Volumetric Meter 

Hot water consumption can be measured using the following three instruments connected together:

    a. Volumetric meter Aquadis+ mark ITRON 
    b. Cyble Impulse Sensor mark ITRON
    c. OGGA Impulse Counter mark OGGA 

The Aquadis + volumetric meter is directly connected to the hot water pipes at the end of the cumulus. This volumetric meter works on the principle of the piston placed inside it. Each rotation of the piston in the measuring chamber represents a given volume of water passing through. At the top of the volumetric meter is placed a cyble impulse sensor, responsible for transmitting a signal to the impulse counter for each turn of the volume meter. Once the information is received from the impulse counter, the instrument then transmits it to the Eco-Touch® device via the EnOcean® radio protocol.

## Retrieval of information

All measurements, direct and indirect, are received from the Eco-Touch® device and then sent to the OGGA company via a 4G modem. OGGA then sends the data to us as a compressed file when requested.
